```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/images.jpeg", 0)  # grayscale
equalized = cv2.equalizeHist(img)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.title("Original"); plt.imshow(img, cmap="gray")
plt.subplot(1,2,2); plt.title("Histogram Equalized"); plt.imshow(equalized, cmap="gray")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/images.jpeg", 0)

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl_img = clahe.apply(img)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.title("Original"); plt.imshow(img, cmap="gray")
plt.subplot(1,2,2); plt.title("CLAHE Enhanced"); plt.imshow(cl_img, cmap="gray")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/images.jpeg")
blur = cv2.GaussianBlur(img, (5,5), 0)

plt.subplot(1,2,1); plt.title("Original"); plt.imshow(img[...,::-1])
plt.subplot(1,2,2); plt.title("Gaussian Blur"); plt.imshow(blur[...,::-1])
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("/content/images.jpeg")

kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])

sharpened = cv2.filter2D(img, -1, kernel)

plt.subplot(1,2,1); plt.title("Original"); plt.imshow(img[...,::-1])
plt.subplot(1,2,2); plt.title("Sharpened"); plt.imshow(sharpened[...,::-1])
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/images.jpeg", 0)
lap = cv2.Laplacian(img, cv2.CV_64F)

plt.subplot(1,2,1); plt.title("Original"); plt.imshow(img, cmap="gray")
plt.subplot(1,2,2); plt.title("Laplacian Edges"); plt.imshow(lap, cmap="gray")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/images.jpeg")

alpha = 1.5  # contrast
beta = 40    # brightness

adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

plt.subplot(1,2,1); plt.title("Original"); plt.imshow(img[...,::-1])
plt.subplot(1,2,2); plt.title("Brightness/Contrast"); plt.imshow(adjusted[...,::-1])
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# -------- Read Image ----------
img = cv2.imread("/content/images.jpeg", 0)   # Load grayscale image

# -------- Apply Histogram Equalization ----------
hist_eq = cv2.equalizeHist(img)

# -------- Apply CLAHE ----------
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_img = clahe.apply(img)

# -------- Plot Histograms ----------
plt.figure(figsize=(12,6))

plt.hist(img.ravel(), 256, [0,256], label="Original", alpha=0.6)
plt.hist(hist_eq.ravel(), 256, [0,256], label="Histogram Equalized", alpha=0.6)
plt.hist(clahe_img.ravel(), 256, [0,256], label="CLAHE", alpha=0.6)

plt.title("Histogram Comparison of Image Enhancement Techniques")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```
